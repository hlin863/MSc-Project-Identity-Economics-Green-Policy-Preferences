```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr
```

```python
# 0.5018	1.3322
# 0.6810	1.6094
# 0.4396	1.6094
# 0.4338	1.3322
# 0.6400	1.6094
# 0.3533	1.3863
# 0.5198	0.6931
# 0.7417	0.6931
# 0.4792	1.6094
# Define lists for cosine similarity and mutual information scores
cosine_similarity = [0.5018, 0.6810, 0.4396, 0.4338, 0.6400, 0.3533, 0.5198, 0.7417, 0.4792]
mutual_information = [1.3322, 1.6094, 1.6094, 1.3322, 1.6094, 1.3863, 0.6931, 0.6931, 1.6094]
```

```python
# Compute Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr(cosine_similarity, mutual_information)
```

```python
# Print the correlation coefficient and p-value
print(f"Pearson correlation coefficient: {correlation_coefficient:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=cosine_similarity, y=mutual_information, marker='o', s=100)

# Adding a regression line
sns.regplot(x=cosine_similarity, y=mutual_information, scatter=False, color='red')

plt.title('Correlation between Cosine Similarity and Mutual Information')
plt.xlabel('Cosine Similarity')
plt.ylabel('Mutual Information')
plt.grid(True)
plt.show()
```
